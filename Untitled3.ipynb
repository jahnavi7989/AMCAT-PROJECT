```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_excel('Amcat.xlsx')

# Display the first few rows, shape, and description
print(data.head())
print(f"Shape: {data.shape}")
print(data.describe())
```

```python
# Identify missing values marked as -1
missing_values = (data == -1).sum()
```

```python
# Plot missing values
plt.figure(figsize=(10,6))
sns.barplot(x=missing_values.index, y=missing_values.values)
plt.xticks(rotation=90)
plt.title('Missing Values per Column')
plt.ylabel('Count of Missing Values')
plt.show()
```

```python
# Plot the distribution of Salary
plt.figure(figsize=(8, 6))
sns.histplot(data['Salary'], kde=True, bins=30, color='blue')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot for Salary to check for outliers
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['Salary'], color='lightgreen')
plt.title('Boxplot of Salary')
plt.show()
```

```python
# List of numerical columns for univariate analysis
numerical_columns = ['10percentage', '12percentage', 'collegeGPA', 
                     'conscientiousness', 'agreeableness', 'extraversion', 
                     'nueroticism', 'openess_to_experience']

# Plot distributions for numerical columns
plt.figure(figsize=(16, 12))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(4, 2, i)
    sns.histplot(data[col], kde=True, bins=30, color='purple')
    plt.title(f'Distribution of {col}')
    plt.tight_layout()

plt.show()
```

```python
# Scatter plot to visualize the relationship between 'Salary' and other numerical columns
plt.figure(figsize=(8, 6))
sns.scatterplot(x='10percentage', y='Salary', data=data)
plt.title('Salary vs 10th Percentage')
plt.show()
```

```python
# Scatter plot to visualize the relationship between 'Salary' and 'collegeGPA'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='collegeGPA', y='Salary', data=data)
plt.title('Salary vs College GPA')
plt.show()
```

```python
# Plot count of employees by Gender
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=data, palette='coolwarm')
plt.title('Count of Employees by Gender')
plt.show()
```

```python
# Filter data for relevant designations
designations = ['Programming Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer']
filtered_data = data[data['Designation'].isin(designations)]

# Check the average salary for these roles
average_salary = filtered_data['Salary'].mean()
print(f"Average salary for relevant roles: {average_salary}")
```

```python
# Use a crosstab to check gender vs specialization
specialization_columns = ['ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 'CivilEngg']

for col in specialization_columns:
    gender_specialization = pd.crosstab(data['Gender'], data[col])
    print(f"Gender vs {col}:")
    print(gender_specialization)
```
